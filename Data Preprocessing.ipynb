### Filter data for training and testing

```python
# Drop columns
df.drop(['aimp','amud','arnd','asin1','asin2','bed1','bed2','bfo1','bfo2'], axis =1 , inplace = True)
df.head()
```

```python
# Rename columns
df.rename(columns={'adbr':'cc1','adfl':'cc2','bso1':'sen1','bso2':'sen2',
                   'bso3':'sen3','ced1':'sys_tem','cfo1':'cool_tem','cso1':'sur_tem'}, inplace = True)
df.head()
```

```python
# Create training dataset (3 days = 259200 records)
df_train= df.iloc[:259200]
df_train.tail()
```

```python
# Create testing dataset (=20% train dataset = 51840)
df_test = df.iloc[259201:311041]
df_test.tail()
```

```python
# Save as .csv file
df_train.to_csv('train1.csv', index = False)
df_test.to_csv('test1.csv', index = False )
```
